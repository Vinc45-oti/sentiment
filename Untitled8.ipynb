```python
import matplotlib.pyplot as plt

# Plot the data
x_train_id = x_train[:, 5]
plt.figure(figsize=(10, 5))
plt.xlabel("Average Number of Rooms")
plt.ylabel("Median Price [$k]")
plt.grid()

plt.scatter(x_train_id, y_train)
plt.show()

```

```python
# Create the model
model = Sequential()
model.add(Dense(units=1, input_shape=(1,)))

# Print model summary
model.summary()
```

```python
model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.005),loss="mse")
```

```python
    history = model.fit(
    x_train_id,
    y_train,
    batch_size=32,
    epochs=101,
    validation_split=0.3
)
```

```python
def plot_loss(history):
    plt.figure(figsize=(20, 5))
    plt.plot(history.history["loss"], 'g', label="Training Loss")
    plt.plot(history.history["val_loss"], 'b', label="Validation Loss")
    plt.xlim([0, 100])  # Fixed typo: xlin → xlim
    plt.ylim([0, 300])  # Fixed typo: ylin → ylim
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid()

plot_loss(history)
```
