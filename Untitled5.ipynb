```python
# Load dataset
df = pd.read_csv("amazon.csv")
df.head()



```

```python
# Kcolumns needed
df = df[['product_name', 'category', 'rating', 'review_content']]
# Preview
print(df.head())
```

```python
# Initialize VADER
analyzer = SentimentIntensityAnalyzer()
# Apply sentiment analysis to each review
df['compound_score'] = df['review_content'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
# Preview results
print(df[['review_content', 'compound_score']].head())
```

```python
# List of aspect keywords to check
aspect_keywords = ["battery", "screen", "camera", "performance", "design", "price", "quality", "sound", "display", "charging"]

# Count how many reviews mention each aspect
aspect_counts = {aspect: df['review_content'].str.contains(aspect, case=False).sum() for aspect in aspect_keywords}

# Display results
for aspect, count in aspect_counts.items():
    print(f"{aspect}: {count} reviews")
```
